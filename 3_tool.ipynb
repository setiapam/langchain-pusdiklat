```python
# Memuat API key dari file .env
from dotenv import load_dotenv
load_dotenv()
```

```python
# Menggunakan pre-built agent dari LangChain untuk menggunakan tools eksternal
from langchain.agents import create_agent
from langchain.messages import HumanMessage

# Memuat tools bawaan dari langchain_community (dalam hal ini, Wikipedia)
from langchain_community.agent_toolkits.load_tools import load_tools
tools = load_tools(["wikipedia"])

# Membuat instance agent dengan model LLM dan daftar tools yang tersedia
agent = create_agent(tools=tools, model="google_genai:gemini-3.1-flash-lite")

# Menyusun pertanyaan pengguna
question = HumanMessage(content="Siapa presiden Indonesia sekarang?")

# Memanggil agent untuk menjawab pertanyaan
# Agent akan memutuskan secara mandiri apakah perlu menggunakan tool Wikipedia
response = agent.invoke(
    {"messages": [question]}
)

print(response)
```
