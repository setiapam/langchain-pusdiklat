```python
# Memuat API key dari file .env
from dotenv import load_dotenv
load_dotenv()
```

```python
# Menggunakan pre-built agent dari LangChain untuk menggunakan tools eksternal
from langchain.agents import create_agent
from langchain.messages import HumanMessage
from langchain_tavily import TavilySearch

# Memuat tool pencarian Tavily menggunakan paket integrasi mandiri yang direkomendasikan
tools = [TavilySearch(max_results=3)]

# Membuat instance agent dengan model LLM dan daftar tools yang tersedia
agent = create_agent(tools=tools, model="google_genai:gemini-3.1-flash-lite")

# Menyusun pertanyaan pengguna tentang final Piala Dunia FIFA 2026
question = HumanMessage(content="Siapa saja tim yang masuk ke babak final Piala Dunia FIFA 2026?")

# Memanggil agent untuk menjawab pertanyaan
# Agent akan menggunakan tool TavilySearch untuk mencari hasil terbaru di internet
response = agent.invoke(
    {"messages": [question]}
)

print(response['messages'][-1].content)
```

```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage

# Membuat instance agent dengan model LLM tanpa mendaftarkan tools
agent = create_agent(model="google_genai:gemini-3.1-flash-lite")

# Menyusun pertanyaan pengguna
question = HumanMessage(content="kapan batasan data yang kamu pakai untuk training?")

# Memanggil agent untuk menjawab pertanyaan
# Agent akan menjawab langsung karena tidak ada tools yang didaftarkan
response = agent.invoke(
    {"messages": [question]}
)

print(response['messages'][-1].content)
```
