## Chat sederhana

```python
# Memuat variabel lingkungan dari file .env (seperti GOOGLE_API_KEY)
from dotenv import load_dotenv

load_dotenv()
```

```python
# Menginisialisasi model chat menggunakan wrapper dari LangChain
# Di sini kita menggunakan model 'gemini-3.1-flash-lite' dari Google GenAI
from langchain.chat_models import init_chat_model

model = init_chat_model("google_genai:gemini-3.1-flash-lite")
```

```python
# Memanggil model dengan mengirimkan pertanyaan dalam bentuk teks biasa
response = model.invoke("Siapa presiden Indonesia yang dijuluki dengan Bapak Pluralisme?")

response
```

```python
# Mencetak isi teks (konten) dari respons yang diberikan oleh model
print(response.content)
```

```python
# Menampilkan metadata respons (seperti nama model, token yang digunakan, dll.)
from pprint import pprint

pprint(response.response_metadata)
```

## Menggunakan Parameter

```python
# Menginisialisasi model dengan konfigurasi parameter tambahan:
# - temperature: mengontrol kreativitas/keacakan respons (nilai lebih tinggi = lebih kreatif)
# - max_tokens: batas maksimum token output yang dihasilkan
# - max_retries: jumlah percobaan ulang otomatis jika terjadi kesalahan koneksi/API
model = init_chat_model(
    model="google_genai:gemini-3.1-flash-lite",
    temperature=1.0,
    max_tokens=None,
    timeout=None,
    max_retries=2
)

response = model.invoke("Siapa presiden Indonesia yang dijuluki dengan Bapak Pluralisme?")
print(response.content)
```

## Menggunakan messages

```python
# Menyusun pesan menggunakan class Message yang terstruktur dari LangChain:
# - SystemMessage: memberikan instruksi dasar tentang bagaimana model harus berperilaku/berperan
# - HumanMessage: pesan dari pengguna yang ingin dijawab oleh model
from langchain.messages import HumanMessage, SystemMessage
model = init_chat_model(
    model="google_genai:gemini-3.1-flash-lite",
    temperature=1.0,
    max_tokens=None,
    timeout=None,
    max_retries=2
)

messages = [
    SystemMessage("Kamu adalah ahli sejarah Indonesia yang ramah dan runtut dalam menjelaskan"),
    HumanMessage("Siapa presiden Indonesia yang dijuluki dengan Bapak Pluralisme?")
]

response = model.invoke(messages)
print(response.content)
```
